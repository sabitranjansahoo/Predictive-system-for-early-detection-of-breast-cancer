```python
from PIL import Image
import os

# Set directory paths
pgm_dir = '/content/drive/MyDrive/major/all-mias'
jpg_dir = '/path/to/save/jpg/images/'

# Loop through PGM files in directory
for filename in os.listdir(pgm_dir):
    if filename.endswith('.pgm'):
        # Open PGM image and save as JPG
        pgm_image = Image.open(os.path.join(pgm_dir, filename))
        jpg_image = pgm_image.convert('RGB')
        jpg_image.save(os.path.join(jpg_dir, filename.replace('.pgm', '.jpg')))
```
