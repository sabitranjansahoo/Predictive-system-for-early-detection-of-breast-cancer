```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file into a Pandas dataframe
data = pd.read_csv('output_final.csv')

# Count the number of occurrences of each class
bg_counts = data['BG'].value_counts()

# Create a bar plot of BG column
fig, ax = plt.subplots(figsize=(8, 6))
bg_counts.plot(kind='bar', ax=ax)

# Set the x-axis label and title
ax.set_xlabel('Character of Background Tissue')
ax.set_ylabel('Count')
ax.set_title('Character of Background Tissue Bar Plot')

# Add the count of every bar to the plot
for i in ax.containers:
    ax.bar_label(i, label_type='edge', padding=2)

# Add a text box to the plot
ax.text(2.8, 80, 'D - Dense-glandular', fontsize=15, color='darkblue')
ax.text(2.8, 60, 'F - Fatty', fontsize=15, color='darkblue')
ax.text(2.8, 40, 'G - Fatty-glandular', fontsize=15, color='darkblue')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file into a Pandas dataframe
data = pd.read_csv('output_final.csv')

# Count the number of occurrences of each class
class_counts = data['CLASS'].value_counts()

# Create a bar plot of BG column
fig, ax = plt.subplots(figsize=(8, 6))
class_counts.plot(kind='bar', ax=ax)

# Set the x-axis label and title
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Bar Plot')

# Add the count of every bar to the plot
for i in ax.containers:
    ax.bar_label(i, label_type='edge', padding=2)

# Add a text box to the plot
plt.text(7, 175, 'NORM - Normal', fontsize=15, color='darkblue')
plt.text(7, 150, 'CALC - Calcification', fontsize=15, color='darkblue')
plt.text(7, 125, 'CIRC - Well-defined/circumscribed masses', fontsize=15, color='darkblue')
plt.text(7, 100, 'ARCH - Architectural distortion', fontsize=15, color='darkblue')
plt.text(7, 75, 'SPIC - Spiculated masses', fontsize=15, color='darkblue')
plt.text(7, 50, 'MISC - Other, ill-defined masses', fontsize=15, color='darkblue')
plt.text(7, 25, 'ASYM - Asymmetry', fontsize=15, color='darkblue')

# Display the plot
plt.show()
```
